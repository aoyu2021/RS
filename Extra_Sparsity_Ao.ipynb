### 1. Ass1 Table

```python
import pandas as pd
import numpy as np
from numpy import count_nonzero

df = pd.read_csv('Global_Baseline.csv')
df
```

```python
df = df.drop(columns = ['Critic'])
df
```

```python
A = np.array(df)

#If you have Nan
A = np.nan_to_num(A,0)

print(A)


# calculate sparsity
sparsity = 1.0 - ( count_nonzero(A) / float(A.size) )
print(sparsity)
```

### 2. jester-data-1 :  Missing Data

```python
df = pd.read_csv('jester-data-1.csv',header = None)
df
```

```python
# replace 99 to 0
df.replace({99:0},inplace=True)
A = np.array(df)

#If you have Nan
A = np.nan_to_num(A,0)
print(A)


# calculate sparsity
sparsity = 1.0 - ( count_nonzero(A) / float(A.size) )
print(sparsity)
```

### 3. Movie Table: Long to Wide

```python
df = pd.read_csv('ratings.csv')
df
```

```python
df = df.drop(columns = ['timestamp'])
df
```

```python
df = df.pivot(index = "userId",columns = "movieId",values = "rating")
df
```

```python
A = np.array(df)

#If you have Nan
A = np.nan_to_num(A,0)
print(A)


# calculate sparsity
sparsity = 1.0 - ( count_nonzero(A) / float(A.size) )
print(sparsity)
```
