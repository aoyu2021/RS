```python
import pandas as pd
import numpy as np


df = pd.read_csv('https://github.com/aoyu2021/RS/blob/main/Global_Baseline.csv')
df
```

```python
df['avg_p'] = df.mean(axis=1)
df
```

```python
df.loc['S_avg'] = df.mean(axis=0)
df
```

```python
df['p_diff'] = df['avg_p'] - df.loc['S_avg','avg_p']
df
```

```python
df = df.T
df
```

```python
df['S_diff'] = df['S_avg'] - df.loc['avg_p','S_avg']
df
```

```python
df = df.T
df
```

```python
# Fill the NaN with 0
df.fillna(0, inplace=True)
df
```

```python
# Loop for replace the 0 to Global Baseline Estimate
for r in df.index:
    for c in df.columns[1:]:
        if df.loc[r, c] == 0:
            df.loc[r, c] = (
                df.loc['S_avg','avg_p']
                + df.loc[r, 'p_diff']
                + df.loc['S_diff', c]
            )
```
